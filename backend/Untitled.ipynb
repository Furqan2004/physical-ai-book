```python
from openai import AsyncOpenAI
import os
from dotenv import load_dotenv

load_dotenv()

# Initialize OpenRouter client for embeddings
embeddings_client = AsyncOpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY"),
)
```

```python
import os
from google import genai
from google.genai import types

# Initialize the Gemini Client
# Note: Ensure GOOGLE_API_KEY is set in your environment
client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY"))

async def get_gemini_embedding(text: str) -> list[float]:
    if not text.strip():
        # Matching your requested dimension for empty inputs
        return [0.0] * 1024  

    try:
        response = await client.aio.models.embed_content(
            model="gemini-embedding-001", 
            contents=text,
            config=types.EmbedContentConfig(
                task_type="RETRIEVAL_DOCUMENT",
                # This explicitly forces the model to output a 1024-length vector
                output_dimensionality=1024
            )
        )
        
        if response and response.embeddings:
            return response.embeddings[0].values
        return None
            
    except Exception as e:
        print(f"❌ Gemini API error: {e}")
        return None

# --- Quick Test ---
# test_vector = await get_gemini_embedding("Testing 1024 dimensions")
# print(len(test_vector)) # Should output 1024
```

```python
sample_text = "Google Gemini embeddings are state-of-the-art for RAG applications."

# Await the async function directly
vector = await get_gemini_embedding(sample_text)

if vector:
    print(f"✅ Success! Vector Dimensions: {len(vector)}")
    print(f"First 5 values: {vector[:5]}")
```
